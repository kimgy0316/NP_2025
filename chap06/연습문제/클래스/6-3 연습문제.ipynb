```python
##연습문제 6-3

## Moneybox 클래스를작성하라. 이클래스는저축한총액을저장하는인스턴스변수가있다. 
## deposit(money) 메서드는 인수로 주어진금액만큼 저축하고 총액을 출력한다, 
## extract(money) 메소드는 money를 인출하고 총액을 출력한다. get_total() 메소드는 총액을
## 알려준다.


class Moneybox:
    def __init__(self):
        self.total = 0

    def deposit(self, money):
        self.total += money
        print(f"{money}원을 저축했습니다. 현재 총액: {self.total}원")

    def extract(self, money):
        self.total -= money
        print(f"{money}원을 인출했습니다. 현재 총액: {self.total}원")

    def get_total(self):
        print(f"현재 총액은 {self.total}원입니다.")
        return self.total

```

```python
## 연습문제 6-3 실행코드

## my_money = Moneybox() 객체를 생성하고 1000원을 저축하고, 200원을 인출한 다음, 총액을 출력하라.

my_money = Moneybox()
my_money.deposit(1000)
my_money.extract(200)
my_money.get_total()
```

```python
## 연습문제 6-4 

## Password_manager 클래스를 작성하라. 이 클래스는 사용자의 모든 과거 패스워드를 저장하는 old_passwords 리스트 변수를 갖는다. 리스트의 마지막 요소는 현재 패스워드이다. 현재 패
## 스워드를 알려주는 get_password와 새로운 패스워드를 설정하는 s et_password 메소드를 정
## 의하라. set_password는 과거에 사용하지 않은 새로운 패스워드를 설정할 때만 유효하다. 문자
## 열을 입력받아 현재 패스워드와 같은지 True 또는 False를 반환하는 is_correct 메소드를 정의하라.

class Password_manager:
    def __init__(self):
        self.old_passwords = []
    def get_password(self):
        """현재 비밀번호 반환"""
        if self.old_passwords:
            return self.old_passwords[-1]
        else:
            return None

    def set_password(self, new_pw):
        """과거에 사용하지 않은 비밀번호만 추가"""
        if new_pw not in self.old_passwords:
            self.old_passwords.append(new_pw)
            print("새로운 비밀번호가 설정되었습니다.")
        else:
            print("이미 사용했던 비밀번호입니다. 다른 비밀번호를 입력하세요.")

    def is_correct(self, test_pw):
        """입력값이 현재 비밀번호와 일치하면 True, 아니면 False"""
        return test_pw == self.get_password()
```

```python
## 연습문제 6-5

## Articles 클래스를 작성하라. 이 클래스는 각각 제품의 이름, 재고수량, 단가를 나타내는 name, stock, price 인스턴스 변수를 가지고 있다. 
## offer_price() 메소드는 구매 수량을 인수로 전달받아 총 가격을 반환한다. 
## 구매수량이 1~10개 이하이면 정상가격을 반환하고, 11~50개이면 5% 할인된 가격을 반환한다. 또 51개 이상이면 10% 할인해준다. purchase() 메소드는
## 구매수량을 인수로 전달받아 재고수량을 구매수량만큼 감소한다.

class Articles:
    def __init__(self, name, stock, price):
        self.name = name
        self.stock = stock
        self.price = price

    def offer_price(self, qty):
        total = self.price * qty
        if 1 <= qty <= 10:
            return total
        elif 11 <= qty <= 50:
            return int(total * 0.95)
        else:  # qty >= 51
            return int(total * 0.90)

    def purchase(self, qty):
        self.stock -= qty
```

```python
## 연습문제 6-5

## 이 클래스를 이용하여 단가 1000원인 ”mask“를 10,000장으로 설정하고, 40장 구매가격과 구매 후 재고를 출력하라.

item = Articles("mask", 10_000, 1000)
price_40 = item.offer_price(40)
item.purchase(40)

print(price_40)
print(item.stock)
```

```python
## 연습문제 6-6

##  다음 프로그램을 실행했을 때 아래 내용이 출력되도록 클래스를 정의하라. 학번(id) 앞 4자리가
## 2018보다 작으면grade() 메소드에의해 “3학년입니다“가 출력되고 아니면 ”3학년이 아닙니다“가 출력된다.

class Student:
    def __init__(self, name, id, dept):
        self.name = name
        self.id = id
        self.dept = dept

    def grade(self):
        year = int(str(self.id)[:4])
        if year < 2018:
            print("3학년입니다")
        else:
            print("3학년이 아닙니다")


man = Student(name="홍길동", id=20150001, dept="로봇공학과")
print("이름: {}, 학번: {}, 학과: {}".format(man.name, man.id, man.dept))
man.grade()
```

```python
## 연습문제 6-9

## 파이썬언어에서다음에대해조사하고예제코드를실행하여보라
## 1. 파이썬으로 싱글톤 패턴을구현하는방법들

## 1)__new__ 오버라이드-------------------------------
class SingletonNew:
    _instance = None
    def __new__(cls, *a, **kw):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance

s1 = SingletonNew(); s2 = SingletonNew()
print(id(s1), id(s2))




## 2) 메타클래스-----------------------------------------
class SingletonMeta(type):
    _instances = {}
    def __call__(cls, *a, **kw):
        if cls not in cls._instances:
            cls._instances[cls] = super().__call__(*a, **kw)
        return cls._instances[cls]

class AppConfig(metaclass=SingletonMeta):
    pass

a = AppConfig(); b = AppConfig()
print(id(a), id(b))





## 3) 데코레이터-------------------------------------------
def singleton(cls):
    instances = {}
    def get_instance(*a, **kw):
        if cls not in instances:
            instances[cls] = cls(*a, **kw)
        return instances[cls]
    return get_instance

@singleton
class Cache:
    def __init__(self): self.data = {}

c1 = Cache(); c2 = Cache()
print(id(c1), id(c2))



## 4) Borg/Monostate(인스턴스는 여러 개, 상태는 공유)------------------------
class Borg:
    _shared = {}
    def __init__(self): self.__dict__ = self._shared

x = Borg(); y = Borg()
x.a = 1
print(y.a)
```

```python
## 연습문제 6-9

## 파이썬언어에서다음에대해조사하고예제코드를실행하여보라
## 2. 파이썬에서 멀티쓰레드구현하는기법

## 1) threading.Thread (target 함수)-----------------------------------
import threading, time

def worker(n):
    print(f"start {n}")
    time.sleep(0.1)
    print(f"end {n}")

threads = [threading.Thread(target=worker, args=(i,)) for i in range(3)]
[t.start() for t in threads]
[t.join() for t in threads]




## 2) Thread 서브클래싱-----------------------------------
import threading, time

class MyThread(threading.Thread):
    def __init__(self, idx): super().__init__(); self.idx = idx
    def run(self):
        print(f"run {self.idx}")
        time.sleep(0.1)

ts = [MyThread(i) for i in range(3)]
[t.start() for t in ts]
[t.join() for t in ts]



## 3) concurrent.futures.ThreadPoolExecutor--------------------------------
from concurrent.futures import ThreadPoolExecutor, as_completed
import time

def task(x):
    time.sleep(0.05)
    return x * x

with ThreadPoolExecutor(max_workers=4) as ex:
    futs = [ex.submit(task, i) for i in range(5)]
    for f in as_completed(futs):
        print(f.result())
```

```python
## 연습문제 6-9

## 파이썬언어에서다음에대해조사하고예제코드를실행하여보라
## 3. 멀티쓰레드에서 공유 변수에대한Critical Section 을 보호하는 기법

## 1) Lock으로 원자성 보장------------------------------
import threading

counter = 0
lock = threading.Lock()

def inc(n):
    global counter
    for _ in range(n):
        with lock:
            counter += 1

threads = [threading.Thread(target=inc, args=(10000,)) for _ in range(4)]
[t.start() for t in threads]
[t.join() for t in threads]
print(counter)


## 2) RLock (재진입 락 같은 스레드가 중첩 획득 가능)------------------------
import threading
lock = threading.RLock()

def outer():
    with lock:
        inner()

def inner():
    with lock:
        pass

outer()



## 3) Semaphore (동시 접근 허용 개수 제한)---------------------------
import threading, time
sem = threading.Semaphore(2)
def work(i):
    with sem:
        print("enter", i)
        time.sleep(0.1)
        print("leave", i)

ts = [threading.Thread(target=work, args=(i,)) for i in range(5)]
[t.start() for t in ts]
[t.join() for t in ts]




## 4) Condition (상태 변화 알림)------------------------------
import threading

ready = threading.Event()

def waiter():
    print("wait")
    ready.wait()
    print("go")

def setter():
    ready.set()

t1 = threading.Thread(target=waiter)
t2 = threading.Thread(target=setter)
t1.start(); t2.start(); t1.join(); t2.join()
```
